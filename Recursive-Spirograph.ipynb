```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Disk:
    """
    A rotating wheel connected to a parent, part of the spirograph stack
    """
    
    def __init__(self, parent, radius, angvel, inner=True, store=None, pencil=None):
        """
        Create a disk and initialise state
        """
        
        # Properties
        self.inner  = inner
        self.radius = radius
        self.angvel = angvel
        self.pencil = pencil
        
        # Relative position
        self.body_angle = 0    # Angle of the disk relative to the parent rotor
        self.rotor_angle = 0   # Angle of the rotor relative to the disk
        
        # Attach to parent
        self.child = None
        self.parent = None
        if parent is not None:
            self.parent = parent
            self.parent.attach(self)
        
        # Attach the store object
        self.store = store
            
    
    def attach(self, child):
        """
        Attach child disk
        """
        
        self.child = child
    
    
    def increment(self):
        """
        Increment time
        """
        
        # Update rotor angle
        self.rotor_angle += self.angvel
        
        # Modify child's body angle accordingly and recurse
        if self.child is not None:
            if self.inner:
                self.child.body_angle -= self.angvel * (self.radius/self.child.radius)
            else:
                self.child.body_angle += self.angvel * (self.radius/self.child.radius)
            self.child.increment()
    
    
    def emit_position(self):
        """
        Calculate absolute position and store it
        """
        
        if self.parent is None:
            x = 0;
            y = 0;
            body = self.body_angle;
            rotor = self.rotor_angle;
        else:
            parent_x, parent_y, parent_body, parent_rotor = self.parent.emit_position()
            if self.inner:
                x = parent_x + (self.parent.radius-self.radius) * np.sin(parent_rotor)
                y = parent_y + (self.parent.radius-self.radius) * np.cos(parent_rotor)
            else:
                x = parent_x + (self.parent.radius+self.radius) * np.sin(parent_rotor)
                y = parent_y + (self.parent.radius+self.radius) * np.cos(parent_rotor)
            body = parent_rotor + self.body_angle
            rotor = body + self.rotor_angle
        
        if self.pencil is not None:
            pencilx = x + self.pencil * self.radius * np.sin(rotor)
            pencily = y + self.pencil * self.radius * np.cos(rotor)
        
        self.store[self]['x'].append(x)
        self.store[self]['y'].append(y)
        self.store[self]['body'].append(body)
        self.store[self]['rotor'].append(rotor)
        if self.pencil is not None:
            self.store[self]['pencilx'].append(pencilx)
            self.store[self]['pencily'].append(pencily)
        
        
        return x, y, body, rotor
        
    
    
```

```python
N = 4
radius = [1.0, 1.0/2.0, 1.0/4.0, 1.0/8.0]
angvel = [2*np.pi/100, 2*np.pi/200, 2*np.pi/1600, 2*np.pi/400]

disk_list = []
parent = None
store = dict()
for ii in range(N):
    disk = Disk(parent, radius[ii], angvel[ii], store=store)
    disk_list.append(disk)
    store[disk] = dict()
    store[disk]['x'] = []
    store[disk]['y'] = []
    store[disk]['body'] = []
    store[disk]['rotor'] = []
    parent = disk

```

```python
for tt in range(10000):
    disk_list[0].increment()
    disk_list[-1].emit_position()
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect('equal')
for disk in reversed(disk_list):
#for disk in [disk_list[-1]]:
    ax.plot(store[disk]['x'], store[disk]['y'])
plt.show()
```
